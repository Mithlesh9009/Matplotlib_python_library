```python
# Create a 2x2 grid of subplots using subplot(). In each subplot, plot different mathematical functions using plot().
# Each subplot should represent a different mathematical function.
#In the first subplot (top-left), plot a sine wave (y = sin(x)) for x values between 0 and 2π.
#In the second subplot (top-right), plot a cosine wave (y = cos(x)) for x values between 0 and 2π.
#In the third subplot (bottom-left), plot a parabola (y = x²) for x values between -5 and 5.
#In the fourth subplot (bottom-right), plot a linear function (y = 2x + 1) for x values between -5 and 5.

x = np.arange(0,np.pi*2)
x1 = np.linspace(0, 2*np.pi, 100)
x2 = np.linspace(-5, 5, 100)
fig, axs = plt.subplots(2, 2, figsize=(10, 8))
axs[0, 0].plot(x1, np.sin(x1), color="b")
axs[0, 0].set_title("Sine Wave (y = sin(x))")
axs[0, 1].plot(x1, np.cos(x1), color="r")
axs[0, 1].set_title("Cosine Wave (y = cos(x))")
axs[1, 0].plot(x2, x2**2, color="g")
axs[1, 0].set_title("Parabola (y = x²)")
axs[1, 1].plot(x2, 2*x2 + 1, color="m")
axs[1, 1].set_title("Linear Function (y = 2x + 1)")
plt.tight_layout()
plt.show()
```
